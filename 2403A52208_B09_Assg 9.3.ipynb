Import Libraries

```python
import numpy as np #For vector operations and numerical computations
import pandas as pd
import gensim #To load and use word embedding models like Word2Vec
from gensim.models import keyedvectors #To load pre-trained embeddings efficiently
from sklearn.metrics.pairwise import cosine_similarity #to compute similarity between word vectors
import matplotlib.pyplot as plt#To visualize word embeddings
```

Load Word2Vec and find numerical vector representation of words

```python
import gensim.downloader as api
from gensim.models import KeyedVectors

# Load pre-trained Word2Vec model (may take time on first download)
model = api.load("word2vec-google-news-300")

# Print vocabulary size
print("Vocabulary Size:", len(model.key_to_index))

# Display vector for a sample word
word = "apple"
vector = model[word]

print("\nWord:", word)
print("Vector length:", len(vector))
print("First 10 values of the vector:\n", vector[:10])
```

Load Word2Vec and complete word similarity


```python
import gensim.downloader as api

# Load pre-trained GloVe model (100D)
model = api.load("glove-wiki-gigaword-100")

# Define word pairs
word_pairs = [
    ("doctor", "nurse"),
    ("cat", "dog"),
    ("car", "bus"),
    ("king", "queen"),
    ("man", "woman"),
    ("teacher", "student"),
    ("apple", "banana"),
    ("computer", "keyboard"),
    ("sun", "moon"),
    ("river", "water")
]

print("Word Similarity Scores:\n")

for w1, w2 in word_pairs:
    similarity = model.similarity(w1, w2)
    print(f"{w1} - {w2} : {similarity:.4f}")
```

Nearest Neighbor Exploration

```python
import gensim.downloader as api

# Load pre-trained GloVe embeddings (100D)
model = api.load("glove-wiki-gigaword-100")

# Choose at least 5 words
chosen_words = ["king", "university", "doctor", "car", "music"]

for word in chosen_words:
    print(f"\nTop similar words for '{word}':\n")

    similar_words = model.most_similar(word, topn=5)

    for similar_word, score in similar_words:
        print(f"{similar_word} : {score:.4f}")
```

Word Analogy Tasks

```python
import gensim.downloader as api

# Load pre-trained Word2Vec (better for analogies)
model = api.load("word2vec-google-news-300")

# Analogy 1
result1 = model.most_similar(
    positive=["king", "woman"],
    negative=["man"],
    topn=5
)

# Analogy 2
result2 = model.most_similar(
    positive=["paris", "india"],
    negative=["france"],
    topn=5
)

# Analogy 3
result3 = model.most_similar(
    positive=["teacher", "hospital"],
    negative=["school"],
    topn=5
)

print("\nking - man + woman = ?")
print(result1)

print("\nparis - france + india = ?")
print(result2)

print("\nteacher - school + hospital = ?")
print(result3)
```
