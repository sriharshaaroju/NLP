Next Word Prediction Using Bi-Gram Counts

```python
def predict_next_word_bigram(word_sequence, bigram_counts, unigram_counts):

    words_in_sequence = word_sequence.lower().split()

    if not words_in_sequence:
        return "Please provide a word sequence."

    last_word = words_in_sequence[-1]

    # Find possible next words
    potential_next_words = {}

    for (w1, w2), count in bigram_counts.items():
        if w1 == last_word:
            potential_next_words[w2] = count

    if not potential_next_words:
        return f"No bigram found starting with '{last_word}'."

    last_word_count = unigram_counts.get(last_word, 0)

    if last_word_count == 0:
        return f"'{last_word}' not found in unigram counts."

    predicted_word = None
    max_probability = 0

    print("\nProbabilities:")

    for next_word, bigram_count in potential_next_words.items():
        probability = bigram_count / last_word_count
        print(f"P({next_word} | {last_word}) = {probability:.3f}")

        if probability > max_probability:
            max_probability = probability
            predicted_word = next_word

    return predicted_word
```

Next Word Prediction Using Trigram Counts

```python
def predict_next_word_trigram(word_sequence, trigram_counts, bigram_counts):

    words_in_sequence = word_sequence.lower().split()

    if len(words_in_sequence) < 2:
        return "Please provide a sequence of at least two words."

    # Extract the last two words to form the prefix bigram
    prefix_bigram = tuple(words_in_sequence[-2:])

    # Find possible next words based on the prefix bigram
    potential_next_words = {}
    for (w1, w2, w3), count in trigram_counts.items():
        if (w1, w2) == prefix_bigram:
            potential_next_words[w3] = count

    if not potential_next_words:
        return f"No trigram found starting with '{' '.join(prefix_bigram)}'."

    # Get the count of the prefix bigram for probability calculation
    bigram_count_of_prefix = bigram_counts.get(prefix_bigram, 0)

    if bigram_count_of_prefix == 0:
        # This should ideally not happen if trigrams were found, but good to check
        return f"Prefix bigram '{' '.join(prefix_bigram)}' not found in bigram counts."

    predicted_word = None
    max_probability = 0

    print(f"\nProbabilities for '{' '.join(prefix_bigram)}':")

    for next_word, trigram_count in potential_next_words.items():
        probability = trigram_count / bigram_count_of_prefix
        print(f"P({next_word} | {' '.join(prefix_bigram)}) = {probability:.3f}")

        if probability > max_probability:
            max_probability = probability
            predicted_word = next_word

    return predicted_word
```

Next Word Prediction Using Bi-Gram Counts with Laplace Smoothening

```python
def predict_next_word_bigram_Laplace(word_sequence, bigram_counts):
    # Tokenize the input sequence and get the last word
    words_in_sequence = word_sequence.lower().split()
    if not words_in_sequence:
        return "Please provide a word sequence."
    last_word = words_in_sequence[-1]

    # Find potential next words based on bigrams starting with last_word
    potential_next_words = {}
    for (w1, w2), count in bigram_counts.items():
        if w1 == last_word:
            potential_next_words[w2] = count

    # Calculate denominator for Laplace smoothing
    last_word_unigram_count = unigram_counts.get(last_word, 0)
    denominator = last_word_unigram_count + V

    if denominator == 0: # This case would only happen if V=0 and last_word_unigram_count=0, but good to handle
        return f"Cannot calculate probability for '{last_word}' (denominator is zero)."
    predicted_word = None
    max_probability = -1 # Initialize with a value less than any possible probability

    print(f"\nProbabilities for next word given '{last_word}' (with Laplace smoothing):")

    # Iterate through all unique words in the vocabulary (unigram_counts keys)
    for next_word in unigram_counts.keys():
        # Get bigram count (w1, w2) and apply Laplace Add-1 smoothing
        bigram_count_with_smoothing = bigram_counts.get((last_word, next_word), 0) + 1

        # Calculate smoothed probability
        probability = bigram_count_with_smoothing / denominator

        print(f"P({next_word} | {last_word}) = {probability:.3f}")

        if probability > max_probability:
            max_probability = probability
            predicted_word = next_word

    if predicted_word is None: # Fallback if for some edge case no word was selected
        return f"No next word could be predicted for '{last_word}'."

    return predicted_word
```

Next Word Prediction Using Tri-Gram Counts based on laplace smoothening

```python
def predict_next_word_trigram_Laplace(word_sequence, Trigrams_counts, bigram_counts):
    # Tokenize the input sequence
    words_in_sequence = word_sequence.lower().split()
    if not words_in_sequence:
        return "Please provide a word sequence."

    # Ensure at least two words for trigram prediction
    if len(words_in_sequence) < 2:
        return "Sequence must contain at least two words for trigram prediction."

    # Get the last two words as a tuple
    last_two_words_tuple = tuple(words_in_sequence[-2:])

    # Find potential next words based on trigrams starting with the last two words
    potential_next_words = {}
    for (w1, w2, w3), count in Trigrams_counts.items():
        if (w1, w2) == last_two_words_tuple:
            potential_next_words[w3] = count

    if not potential_next_words:
        return f"No trigram found starting with '{' '.join(last_two_words_tuple)}'."

    # Calculate probabilities for potential next words
    # P(w3 | w1,w2) = Count(w1, w2, w3) / Count(w1, w2)
    # The denominator should be the count of the bigram (w1, w2)
    last_two_words_bigram_count = bigram_counts.get(last_two_words_tuple, 0)
    if last_two_words_bigram_count == 0:
        return f"'{' '.join(last_two_words_tuple)}' not found as a bigram. Cannot predict next word."

    predicted_word = None
    max_probability = -1

    for next_word, trigram_count in potential_next_words.items():
        probability = (trigram_count+1) / (last_two_words_bigram_count+V)
        print("probability of  " f'{next_word}' " is ", probability)
        if probability > max_probability:
            max_probability = probability
            predicted_word = next_word

    return predicted_word
```

Next Word Prediction Using Bi-Gram Counts with Add - K Smoothening

```python
def predict_next_word_bigram_K(word_sequence, bigram_counts, unigram_counts, K): #K=0.5-0.01
    # Tokenize the input sequence and get the last word
    words_in_sequence = word_sequence.lower().split()
    if not words_in_sequence:
        return "Please provide a word sequence."
    last_word = words_in_sequence[-1]

    # Find potential next words based on bigrams starting with last_word
    potential_next_words = {}
    for (w1, w2), count in bigram_counts.items():
        if w1 == last_word:
            potential_next_words[w2] = count

    if not potential_next_words:
        return f"No bigram found starting with '{last_word}'."

    # Calculate probabilities for potential next words
    # P(w2 | w1) = Count(w1, w2) / Count(w1)
    last_word_unigram_count = unigram_counts.get(last_word, 0)
    if last_word_unigram_count == 0:
        return f"'{last_word}' not found in unigram counts. Cannot predict next word."

    predicted_word = None
    max_probability = -1

    for next_word, bigram_count in potential_next_words.items():
        probability = (bigram_count+K) / (last_word_unigram_count+K*V)
        print("probability of " f'{next_word}' " is ", probability)
        if probability > max_probability:
            max_probability = probability
            predicted_word = next_word

    return predicted_word
```
